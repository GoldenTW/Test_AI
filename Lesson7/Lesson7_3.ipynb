```python
cities = ['Taipei', 'New York', 'Tokyo', 'London', 'Paris']

print(cities)
print(cities[-1])

#slice[前面的索引:後面的索引]
# 包含前面的索引，不包含後面的索引
#
print(f"cities[0:2]={cities[0:2]}")    
print(f"cities[:2]={cities[:2]}")  
print(f"cities[:2]={cities[-2:]}") 

```

```python
#迭代列表 inteval

for city in cities:
    print(city)

```
