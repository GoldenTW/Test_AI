```python
# [1,3,5,7,9]
# 傳統方法建立

numbers = []
for i in range(1, 10, 2):
    numbers.append(i)
print(numbers)  
```

```python
# 使用List Comprehension建立

numbers = [i for i in range(1, 10, 2)]
print(numbers)

print([i for i in range(1, 10) if i % 2 == 1])
```

```python
# 使用Dict Comprehension建立

word = "letter"
print({letter:word.count(letter) for letter in word})
print({letter:word.count(letter) for letter in word if letter == 'e'})
print({letter:word.count(letter) for letter in word if word.count(letter) > 1})
```
