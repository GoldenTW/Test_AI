```python
def menu(wine,entree,dessert):
    print(f"Wine: {wine}")
    print(f"Entree: {entree}")
    print(f"Dessert: {dessert}")

menu("Chardonnay","Chicken","Cake")
```

```python
def menu(wine,entree,dessert):
    print(f"Wine: {wine}")
    print(f"Entree: {entree}")
    print(f"Dessert: {dessert}")

menu(entree="Chicken",dessert="Cake",wine="Chardonnay") 

#引述名稱的呼叫可以不照順序
```

```python
#引述名稱有預設值, 要放最後

def menu(wine,entree,dessert="Ice Cream"):
    print(f"Wine: {wine}")
    print(f"Entree: {entree}")
    print(f"Dessert: {dessert}\n")

menu("Chardonnay","Chicken")
menu("Chardonnay","Chicken","Cake")
```
