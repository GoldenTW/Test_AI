```python
#猜數字遊戲

import random

min=1
max=100
target = random.randint(min, max)
print("=========== 猜數字遊戲開始 =========== \n")

while True:
    guess = int(input(f"請輸入一個{min}到{max}之間的數字: "))
    if guess < min or guess > max:
        print(f"超過範圍, 程式終止")
        break
    if guess < target:
        print("太小了!")
    elif guess > target:
        print("太大了!")
    else:
        print(f"恭喜你，猜對了!{target}就是答案!")
        break

```

```python
#猜數字遊戲

import random

def guess_number_game():
    min=1
    max=100
    target = random.randint(min, max)
    print("=========== 猜數字遊戲開始 =========== \n")
    count=0

    while True:
        guess = int(input(f"請輸入一個{min}到{max}之間的數字: "))
        count+=1
        if guess < min or guess > max:
            print(f"超過範圍, 程式終止")
            break
        if guess < target:
            print(f"太小了!你玩了{count}次")
        elif guess > target:
            print(f"太大了!你玩了{count}次")
        else:
            print(f"恭喜你，猜對了!{target}就是答案!你總共玩了{count}次")
            break

while True:
    guess_number_game()
    play_again = input("是否要再玩一次? (y/n): ")
    if play_again.lower() != 'y':
        print("感謝遊玩，再見!")
        break

    
```
