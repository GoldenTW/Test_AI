```python
#小明想存錢買一輛機車，他決定每個月存下固定的金額，請問他每個月需要存多少錢，才能在一年內存到30000元買機車？

deposit_per_month = 0
target_amount = 30000
n=0
while deposit_per_month < target_amount:
    n += 1
    input_amount = int(input("請輸入第{n}個月存款金額: "))
    deposit_per_month += input_amount
    print(f"第 {n} 個月，存了 {input_amount} 元，目前總計存了 {deposit_per_month} 元。")


print(f"恭喜！你已經存夠了。經過{n}個月,總計存了 {deposit_per_month} 元。")   

```

```python
deposit_per_month = 0
target_amount = 30000
n=0
while 1==1:
    n += 1
    input_amount = int(input("請輸入第{n}個月存款金額: "))
    deposit_per_month += input_amount
    print(f"第 {n} 個月，存了 {input_amount} 元，目前總計存了 {deposit_per_month} 元。")
    if deposit_per_month >= target_amount:
        break


print(f"恭喜！你已經存夠了。經過{n}個月,總計存了 {deposit_per_month} 元。")   
```

```python
# 設計一個程式, 讓使用者輸入數值, 只有加總正偶數值, 不加總正其數值, 如果輸入負值 則結束程式並輸出加總結果。
total = 0
while True:
    num = int(input("請輸入一個整數(輸入負值結束程式): "))
    if num < 0:
        break
    elif num % 2 == 0:
        total += num
print("加總結果為:", total)


```

```python
import random

print(random.random())  # 產生0到1之間的隨機浮點數
print(random.randint(1, 10))  # 產生1到10之間的隨機整數
print(random.choice(['apple', 'banana', 'cherry']))  # 從列表中隨機選擇一個元素
print(random.uniform(1.5, 5.5))  # 產生1.5到5.5之間的隨機浮點數

```
