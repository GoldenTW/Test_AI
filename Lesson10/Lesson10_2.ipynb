```python

try: # This block runs first
    raise ValueError
except: # This block runs if an exception was raised
    print("An error occurred")
else: # This block runs if no exception was raised
    print("No errors occurred")
    print("Execution successful")
finally: # This block always runs
    print("Execution completed")

print("Program End...")


```

```python
try:
    money_Temp = input("請輸入購物金額: ")
    money = int(money_Temp)
except:
    print(f"輸入錯誤{money_Temp}，請輸入數字")
else:
    print(f"購物金額為: {money} 元")
finally:
    print("程式結束")

```

```python
#處理錯誤,會讓程式增加可靠度
#Exception可以多個except區塊,由上而下處理滿足的一個

try:
    money_Temp = input("請輸入購物金額: ")
    money = int(money_Temp)
    if money < 0:
        raise Exception ("購物金額不可為負數")
except ValueError: # This block runs for ValueError exceptions
    print(f"輸入錯誤{money_Temp}，請輸入數字")
except Exception as e: # This block runs for other exceptions
    print(e)
else:
    print(f"購物金額為: {money} 元")
finally:
    print("程式結束")
```
