```python
#自訂Clase

class Person:  
    def __init__(self,name:str,age:int):
        self.name = name
        self.age = age

    #實體方法
    def description(self):
        print(f"我的名字叫{self.name}")
        print(f"我的age是{self.age}")
```

```python
p1 = Person("Robet",30)
p1.description()
```
